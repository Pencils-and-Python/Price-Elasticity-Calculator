# 📊 Understanding Elasticity

Elasticity is one of those deceptively simple ideas that packs a ton of practical power. At its core, **elasticity measures how sensitive one thing is to changes in another thing.**

In this project, we’re focused on **Price Elasticity of Demand (PED).**

---

## 💡 The Big Idea

> **If you change the price of a product, how much will the quantity demanded by consumers change in response?**

- If people **stop buying immediately** when the price goes up a little? That’s **elastic.**
- If people **keep buying no matter what** (like medicine)? That’s **inelastic.**

In other words: *How much do buyers care about the price?*

---

## 🧮 The Formula

The classic formula for PED is:

$$
E_d = \frac{\%\ \text{Change in Quantity Demanded}}{\%\ \text{Change in Price}}
$$

**Example:**

If the price of a product goes **up 10%** and demand drops **20%**, then:

$$
E_d = \frac{-20\%}{+10\%} = -2
$$

The **negative sign** is normal because price and demand usually move in opposite directions. Often, we focus on the *absolute value* when classifying elasticity.

---

## 🚦 What the Numbers Mean

| **Elasticity Value**            | **What It Means**                                   |
|----------------------------------|-----------------------------------------------------|
| \( E_d > 1 \) (Elastic)          | Quantity demanded changes **more than** the price.  |
| \( E_d = 1 \) (Unitary Elastic)  | Quantity changes by **exactly the same %** as price.|
| \( E_d < 1 \) (Inelastic)        | Quantity changes **less than** the price.           |

**Shortcut thinking:**
- **Elastic:** Luxury goods, optional stuff.
- **Inelastic:** Necessities, things you can’t easily substitute.

---

## 💰 Why It Matters

The magic of elasticity is in **pricing strategy:**

- **If demand is elastic:** Raising prices = 🚫 bad idea → revenue *shrinks*.
- **If demand is inelastic:** Raising prices = ✅ good idea → revenue *grows* (to a point).

This is crucial for businesses, governments, and even civil infrastructure (like water rates!).

---

## 🌍 Real-World Examples

- **Elastic:**
    - High-end electronics: If an iPhone jumps $500 in price, people *hesitate or switch.*
    - Airline tickets: If one airline jacks up prices, you’ll check competitors.

- **Inelastic:**
    - Medicine: People will pay what they have to.
    - Utilities: Basic services (water, electricity) that people *need.*
